```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.islower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

```python
def third_of_total(text):
    total_words = [w for w in text if any(c.isalpha() for c in w)]

    frdist = nltk.FreqDist(total_words)
    most_common = frdist.most_common(1000)
    count = 0
    third = []

    for w, i in most_common:
        if ((count < (len(total_words) / 3)) & any(c.isalpha() for c in w)):
            #print(i)
            count = count + i
            #print(len(total_words)/3)
            #print(count)
            third.append(w)
    print (third)        
    print(len(third))

third_of_total(bw)
third_of_total(text1)
third_of_total(text5)
third_of_total(text4)
```

```python
#14 ◑ Define a function supergloss(s) that takes a synset s as its argument and returns a string 
#consisting of
#the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.
```

```python
def supergloss(s):
    counthypo = 0
    counthyper = 0
    hyperstring = []
    hypostring = []
    if counthypo < len(s.hyponyms()):
        for hypo in s.hyponyms():
            hypostring.append(str(hypo) + ': ' + hypo.definition())
            counthypo = counthypo + 1
    if counthyper < len(s.hypernyms()):
        for hyper in s.hypernyms():
            hyperstring.append(str(hyper) + ': ' + hyper.definition())
            counthyper = counthyper + 1
    return str(s) + ': ' + s.definition() + ' ' + str(hyperstring) + ' ' + str(hypostring)

supergloss(wn.synset('tree.n.01'))
```

```python
#☼ Save some text into a file corpus.txt. Define a function load(f) that reads from the file named in its sole argument, 
#and returns a string containing the text of the file.
from nltk.corpus import PlaintextCorpusReader
def load(f):
    corpus_root = 'Desktop/'
    wordlists = PlaintextCorpusReader(corpus_root, f)
    wordlists.words()
    string = (' ').join(wordlists.words())
    newstring = nltk.re.sub('\s,', ',', string)
    newstring
    newerstring = nltk.re.sub('\s\.\s|\s\.', '. ', newstring)
    return newerstring


stringified = load('corpus.txt')
stringified
```

```python
#Use one multi-line regular expression, with inline comments, using the verbose flag (?x). Use nltk.regexp_tokenize() 
#to create a tokenizer that tokenizes the following kinds of expression: monetary amounts; dates; names of people 
#and organizations. 

text = 'Aunt Jessica gave me a $50 Amazon gift card for passing my Biology test on 31/07/2018, I passed with 80% on 22-07-2018. US$ 50. I spent it on 28-07. $49.99. 0.99'

pattern = r'''(?x)
    \$\s?(?:\d+\.)?\d+                  # Monetary amount: $50, $49.99
    | \d{2}\-\d{2}\-\d{4}               # Date format: 31-07-2018
    | \d{2}\/\d{2}\/\d{4}               # Date format: 31/07/2018
    | [A-Z][a-z]+                       # Names of people and orgs
     '''
#nltk.re_show('([A-Z][a-z]+)', text)
#nltk.re_show('(\d{1,2})(\/|\-)(\d{1,2})?(\/|\-)?\d{2,4}', text)
#nltk.re_show('\w*?\$\s?\d+(\.\d+)?', text)
nltk.regexp_tokenize(text, pattern)
```
