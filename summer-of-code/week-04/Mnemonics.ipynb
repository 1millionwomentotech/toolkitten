## Actual data includes:
1. Relevant - Pregnant
2. Non-relevant - Not-pregnant


## We classify them as (we assign our judgement as):
1. Positive 
2. Negative


## These interact in 4 ways:

### Based on data:
(a) Relevant
1. TP : pregnant women `correctly judged` **HIT**
2. FN : pregnant women whom we told they aren't pregnant `incorrectly judged` **MISS**

(b) Non-relevant
3. TN : non-pregnant women and men `correctly judged` **CORRECT REJECTION**
4. FP : non-pregnant women and men (!) we told they ARE pregnant `incorrectly judged` **FALSE ALARM**

## Same breakdown can be done based on our judgements:

(a) Correctly judged
1. TP : pregnant women `correctly judged` **HIT**
3. TN : non-pregnant women and men `correctly judged` **CORRECT REJECTION**

(b) Incorrectly judged
2. FN : pregnant women whom we told they aren't pregnant `incorrectly judged` **MISS**
4. FP : non-pregnant women and men (!) we told they ARE pregnant `incorrectly judged` **FALSE ALARM**

# Accuracy Measures

- Accuracy 
- Precision
- Recall
- F-score

## Accuracy (Rand Index, Rand Measure)

TP + TN / Total
HIT & CORRECT REJECTION / Total

## Precision

The stuff you found, how good was it?

TP / P

TP / TP + FP

HIT / HIT & FALSE ALARM

HIT / Positive

## Recall

Did you find it all?

HIT / Relevant

## F-score (F-measure)

2 P * R / (P + R)

2 (HIT / Positive) * (HIT / Relevant) / ( (HIT / Positive) + (HIT / Relevant) )
